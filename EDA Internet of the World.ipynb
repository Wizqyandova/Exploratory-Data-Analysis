## 3. 5 Negara Apa dengan user Internet terbanyak Tahun 2016?

```python
df_2016 = df[df['Year'] == 2016]
```

```python
pd.crosstab(index=df_2016['Entity'], columns='Jumlah Pengguna', values=df_2016['User'], aggfunc='sum').sort_values('Jumlah Pengguna', ascending=False).head()
```

Insight = 5 Negara tersebut memiliki Jumlah pengguna Internet Tertinggi pada Tahun 2016 sehingga dapat dijadikan sebagai pembandingan untuk memilih negara tersebut
